```python
import os
import json

from pprint import pprint
```

```python
train_images_path = './images/train/'
train_labels_path = './images/train.json'
```

```python
# get the names of images in the training images directory
images = sorted(os.listdir(train_images_path))
print(images)
```

```python
# get the training annotations json file
with open(train_labels_path) as data:
    labels = json.load(data)
    pprint(labels)
```

```python
# transform the labels into a dictionary for convenience
labels = {label['image']: label for label in labels}
pprint(labels)
```

```python
# for each image, get and print its question
image_questions = {}
for image in images:
    image_questions[image] = labels[image]['question']
    
pprint(image_questions)
```
