```python
# load MNIST dataset and split into training/testing sets
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets("/tmp/data/")
X_train = mnist.train.images
X_test = mnist.test.images
y_train = mnist.train.labels.astype("int")
y_test = mnist.test.labels.astype("int")
```

```python
from tensorflow.contrib.learn import DNNClassifier, SKCompat, RunConfig, \
                                     infer_real_valued_columns_from_input

# DNN configuration
config = RunConfig(tf_random_seed=42)

# extract features from the training data
feature_columns = infer_real_valued_columns_from_input(X_train)

# instantiate DNN with two hidden layers (300 and 100 neurons) and a
# softmax output layer (10 neurons)
dnn_clf = DNNClassifier(hidden_units=[300,100],
                        n_classes=10,
                        feature_columns=feature_columns,
                        config=config)

# wrapper estimator for compatability (if TensorFlow >= 1.1)
dnn_clf = SKCompat(dnn_clf)

# train DNN using stochastic gradient descent
dnn_clf.fit(X_train, y_train, batch_size=50, steps=20000)
```

```python
from sklearn.metrics import accuracy_score, log_loss

# evaluate accuracy
y_pred = dnn_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred['classes'])

# evaluate loss
y_pred_proba = y_pred['probabilities']
loss = log_loss(y_test, y_pred_proba)

# print results
print('Accuracy:\t{}'.format(accuracy))
print('Loss:\t\t{}'.format(loss))
```
